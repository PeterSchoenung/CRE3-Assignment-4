# Backmixed Plug Flow Tubular Reactors (PFTR): Effect of Reaction Order and Axial Dispersion

---

## 1. Fundamentals of Backmixing in Plug Flow Tubular Reactors

An ideal Plug Flow Tubular Reactor (PFTR) assumes that all fluid elements move through the reactor with the same velocity and experience the same residence time. Under these assumptions, no longitudinal mixing occurs and concentration decreases continuously along the reactor length.

However, real tubular reactors rarely behave ideally. Molecular diffusion, turbulent eddies, velocity gradients, and packing irregularities cause **axial dispersion (backmixing)**, allowing material to move not only downstream by convection but also upstream by diffusion. As a result, fluid elements experience different residence times, leading to a broader Residence Time Distribution (RTD) than predicted by the ideal PFTR model. Danckwerts introduced RTD precisely to describe these deviations from ideal flow and demonstrated that most practical reactors behave somewhere between ideal plug flow and complete mixing [3]. 

The presence of axial dispersion transforms the governing reactor equation into a second-order convection–diffusion–reaction equation

$$
D_{ax}\frac{d^2C}{dz^2}-u\frac{dC}{dz}-r(C)=0
$$

where:

- $D_{ax}$ = axial dispersion coefficient
- $u$ = superficial velocity
- $r(C)$ = reaction rate

---

## 2. Behaviour of a Backmixed PFTR with a First-Order Reaction

For a first-order reaction,

$$
r=kC
$$

the reaction rate is directly proportional to the reactant concentration.

In an ideal PFTR, concentration decreases monotonically from the inlet to the outlet, maintaining relatively steep concentration gradients that continuously drive the reaction forward.

When axial dispersion is introduced, fresh reactant mixes with partially converted fluid moving upstream. This longitudinal mixing smooths the concentration profile throughout the reactor, reducing concentration gradients. Consequently, the average reactant concentration near the reactor outlet becomes higher than in an ideal PFTR, while the overall conversion decreases.

Although first-order kinetics are relatively tolerant to concentration fluctuations, backmixing still lowers reactor efficiency because reactants spend part of their residence time mixed with already converted material instead of continuously encountering fresh reactant. Mears identified axial dispersion as one of the primary reasons laboratory packed-bed reactors often exhibit lower conversions than predicted by ideal plug-flow theory [6].

For moderate dispersion levels, the reactor still behaves similarly to a PFTR, but increasing dispersion gradually shifts reactor behaviour toward complete mixing. Therefore, the first-order reaction provides an excellent benchmark for studying the influence of axial dispersion because it isolates hydrodynamic effects without introducing additional nonlinear kinetic complexity. This is also why the Assignment recommends beginning with first-order kinetics.

## 3. Behaviour of a Backmixed PFTR with Reaction Orders Different from One

When the reaction order differs from one, the influence of axial dispersion becomes much more significant because the reaction rate no longer depends linearly on concentration.

### Reaction orders greater than one (n > 1)

For reactions such as

$$
r=kC^2
$$

or

$$
r=kC^3,
$$

the reaction rate increases rapidly with reactant concentration.

Under ideal plug-flow conditions, the inlet region contains high reactant concentrations and therefore extremely high reaction rates.

Backmixing reduces these high local concentrations by mixing fresh reactant with partially converted fluid. Although this mixing slightly increases concentration farther downstream, the gain in those regions cannot compensate for the loss of reaction rate near the inlet because nonlinear kinetics strongly favour high concentrations.

Consequently, higher-order reactions experience:

- A greater reduction in reaction rate.
- A larger decrease in overall conversion.
- Stronger sensitivity to axial dispersion than first-order reactions.

Mears demonstrated that the minimum reactor length required to neglect axial dispersion increases with reaction order, indicating that higher-order reactions become increasingly susceptible to dispersion effects [6].

Dommeti and Balakotaiah further showed that increasing kinetic nonlinearity makes reactor predictions increasingly dependent on the chosen hydrodynamic model. While different one-parameter models remain nearly equivalent for slow reactions, significant deviations appear as reaction intensity increases [4]. 

---

### Reaction orders smaller than one (n < 1)

For fractional-order reactions, the reaction rate is less sensitive to concentration.

Therefore, smoothing the concentration profile through backmixing has a much smaller influence on reactor performance than for second- or third-order reactions.

Although conversion still decreases because residence times become more broadly distributed, the reduction is generally less pronounced.

This trend follows directly from the mathematical dependence of reaction rate on concentration and is consistent with the nonlinear kinetic analysis presented by Dommeti and Balakotaiah [4].

---

## 4. Influence of the Peclet and Bodenstein Numbers

The intensity of axial dispersion is commonly described using either the Peclet number (Pe) or the Bodenstein number (Bo).

Both compare convective transport with longitudinal dispersion.

Large Peclet (or Bodenstein) numbers indicate convection-dominated transport with very little axial mixing. Under these conditions, the reactor behaves almost identically to an ideal PFTR.

Conversely, small Peclet or Bodenstein numbers indicate that dispersion becomes comparable to convection, producing extensive backmixing and reactor behaviour approaching complete mixing.

Bremer and Turek explain that although European literature usually employs the Bodenstein number while American literature often uses the Peclet number, both represent essentially the same physical balance between convection and dispersion. Their review also discusses the historical development of these dimensionless groups and highlights the need for consistent terminology [2]. 

Mears further demonstrated that dispersion effects become increasingly severe as the Bodenstein number decreases, particularly for laboratory-scale reactors operating at low Reynolds numbers [6].

---

## 5. Comparison between the Axial Dispersion Model (ADM) and the Tanks-in-Series Model (TISM)

Both the Axial Dispersion Model (ADM) and the Tanks-in-Series Model (TISM) describe partially mixed tubular reactors using a single adjustable parameter.

The ADM characterizes non-ideal flow using the Peclet number, whereas the TISM represents the reactor as a cascade of $N$ equal-sized Continuous Stirred Tank Reactors (CSTRs). As $N$ increases, the system approaches ideal plug-flow behaviour; as $N=1$, it becomes a single perfectly mixed reactor [1]. 

Abu-Reesh and Abu-Sharkh compared both models over the entire range from PFTR to CSTR and found that they provide nearly identical predictions when their parameters are related by

$$
Pe = 2(N-1).
$$

The agreement is particularly good at high Peclet numbers and low dimensionless residence times. Although the TISM lacks the clear physical interpretation of the ADM, it is computationally simpler because it requires solving algebraic equations rather than second-order differential equations [1]. 

Dommeti and Balakotaiah confirmed that this equivalence remains valid primarily for slow reactions, whereas increasing Damköhler number causes predictions from different models to diverge [4]. 

---

## 6. Published Scientific Results Suitable for Model Validation

The literature consistently reports several qualitative trends that should also appear in numerical simulations of axial dispersion models.

Increasing axial dispersion (decreasing Peclet/Bodenstein number) reduces reactor conversion because longitudinal mixing decreases concentration gradients and lowers the effective driving force for reaction [6]. 

Increasing reaction order amplifies the influence of backmixing because nonlinear kinetics depend more strongly on local reactant concentration. Consequently, second- and third-order reactions experience larger conversion losses than first-order reactions [6]. 

Increasing reactor length or increasing the Peclet/Bodenstein number reduces the relative importance of axial dispersion, causing reactor behaviour to approach ideal plug flow [6]. 

Finally, for sufficiently slow reactions, both the Axial Dispersion Model and the Tanks-in-Series Model should produce very similar conversion predictions when the relationship

$$
Pe = 2(N-1)
$$

is applied [1][6]. 

---

## 7. Final Remarks

The literature consistently supports the following conclusions:

- Real tubular reactors deviate from ideal PFTR behaviour because axial dispersion broadens the Residence Time Distribution (RTD) and introduces longitudinal mixing [3].
- Backmixing reduces reactor conversion by smoothing concentration gradients along the reactor [6]. 
- Higher-order reactions are considerably more sensitive to axial dispersion than first-order reactions because their reaction rates depend more strongly on local reactant concentration [6]. 
- Large Peclet or Bodenstein numbers correspond to nearly ideal plug-flow behaviour, whereas small values indicate stronger backmixing and reactor behaviour approaching complete mixing [2]. 
- The Axial Dispersion Model (ADM) and the Tanks-in-Series Model (TISM) become theoretically equivalent for slow reactions when their parameters are properly related, making either approach appropriate for validating numerical simulations [1][4].
